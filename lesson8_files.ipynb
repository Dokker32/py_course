<h1>PDF</h1>

```python
import glob
import tabula
import pandas as pd
```

Список всех PDF файлов в рабочей папке

```python
pdf_files = glob.glob('data/*.pdf')
pdf_files
```

Прочитать PDF

```python
pdf_path = "data/data.pdf"

dfs = tabula.read_pdf(pdf_path, stream=True)
# read_pdf returns list of DataFrames
print(len(dfs))
dfs[0]
```

```python
# 3 страница
df = tabula.read_pdf(pdf_path, pages=3, stream=True)
df[0]
```

```python
#pdf как JSON
tabula.read_pdf(pdf_path, output_format="json")
```

Преобразование файла PDF напрямую в файл CSV

```python
data = 'data/data.pdf' 
tabula.convert_into(data, "data/data.csv") 
print("Преобразование в csv успешно")
```

```python
df = pd.read_csv('data/data.csv')
df
```

Сохранение CSV

```python
df['new'] = 12
df.to_csv('data/new_df.csv')
df = pd.read_csv('data/new_df.csv')
df
```
